```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_2019 = pd.read_excel('/Users/jkky/Documents/project_vacation_2022/my_data/data_2019.xls')
df_2020 = pd.read_excel('/Users/jkky/Documents/project_vacation_2022/my_data/data_2020.xls')
df_2021 = pd.read_excel('/Users/jkky/Documents/project_vacation_2022/my_data/data_2021.xls')
df_2019['year'] = 2019
df_2020['year'] = 2020
df_2021['year'] = 2021
df_2020.head()
```

```python
df_2019_1 = df_2019[df_2019['semester']=='1학기']
df_2019_2 = df_2019[df_2019['semester']=='2학기']
df_2020_1 = df_2020[df_2020['semester']=='1학기']
df_2020_2 = df_2020[df_2020['semester']=='2학기']
df_2021_1 = df_2021[df_2021['semester']=='1학기']
df_2021_2 = df_2021[df_2021['semester']=='2학기']

df_2021_2.head()
```

```python
class Changer:
    def __init__(self):
        None
    def score_nomalization(self,df):
        df['A0rate'] = df['A0rate'] + df['A-rate']
        df['B0rate'] = df['B0rate'] + df['B-rate']
        df['C0rate'] = df['C0rate'] + df['C-rate']
        df['D0rate'] = df['D0rate'] + df['D-rate']
        
        df_final=df.drop(['A-rate','B-rate','C-rate','D-rate'], axis=1)
        return df_final
```

```python
tmp = Changer()
df_2019_1 = tmp.score_nomalization(df_2019_1)
df_2019_2 = tmp.score_nomalization(df_2019_2)
df_2020_1 = tmp.score_nomalization(df_2020_1)
df_2020_2 = tmp.score_nomalization(df_2020_2)
df_2021_1 = tmp.score_nomalization(df_2021_1)
df_2021_2 = tmp.score_nomalization(df_2021_2)
```

```python
df_2020_1.head()
```

```python
df_total = pd.concat([df_2019_1,df_2019_2,df_2020_1,df_2020_2,df_2021_1,df_2021_2], ignore_index=True)
df_total.count()
```

```python
df_total[df_total['year']==2019]
```

```python
df_total['A'] = df_total['A+rate'] + df_total['A0rate']
```

```python
df_group = df_total.groupby(['year','semester']).agg(np.mean)
```
