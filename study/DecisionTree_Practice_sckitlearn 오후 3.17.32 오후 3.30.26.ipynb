# Train DecisionTree Classifier

Scikit-learn URL :
https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

image.png

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
X = df[['petal length (cm)', 'petal width (cm)']]
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=42)
tree_clf.fit(X, y)

# max_depth : 트리의 깊이를 나타내며 과최적화를 잘 판단하여 선택하여야 한다.
# random_state : 확률변화 고정
# min_sample_split : maxdepth와 비슷한 기능으로 숫자를 부여받고 예로 2로 설정할 경우 2개 이하의 데이터를 가진 
# 노드에서는 더 이상 학습을 진행하지 않게된다.
# max_feature : 많은 특성이 있을 경우 최대 n가지만 사용하도록 제한하는 경우
# criterion : 불순도 계산 방법 정의 (지니계수로 선정)
```

```python
tree_clf.predict_proba([[5, 1.5]])

#predict_proba는 주어진 테스트 예시에 대해 레이블별 속할 확률을 알려줍니다
#predict는 주어진 테스트 예시에 대해 가장 높은 확률의 예측값을 도출합니다.
```

```python
tree_clf.predict([[5, 1.5]])
```

```python
tree_clf.feature_importances_

#모델의 예측에 중요한 칼럼의 기여도를 나타냅니다.
```

# Wine data에 실습

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
data = load_wine(as_frame=True)
df = data.data
df['label'] = data.target

X = df.loc[:, ~df.columns.isin(['label'])]
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 11)

clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)

data.data.head()
```

```python
clf.fit(X_train, y_train)
accuracy_score(y_train, clf.predict(X_train))
```

```python
export_graphviz(
        clf,
        out_file=os.path.join(IMAGES_PATH, "wine.dot"),
        feature_names=data.feature_names,
        class_names=data.target_names,
        rounded=True,
        filled=True
    )
Source.from_file(os.path.join(IMAGES_PATH, "wine.dot"))
```

```python
y_pred = clf.predict(X_test)
print('정확도 : {}'.format(accuracy_score(y_test, y_pred)))
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x = data.feature_names, y = clf.feature_importances_)
plt.xticks(rotation = 90)
```

```python
#flavanoids, color_intensity select

df = data.data[['flavanoids', 'color_intensity']]
df['label'] = data.target

X = df[['flavanoids', 'color_intensity']]
y = df['label']

X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=40)
```

```python
clf_2f = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=40)
clf_2f.fit(X_train, Y_train)
clf_2f.predict(X_train)
accuracy_score(clf_2f.predict(X_train), Y_train)
```

```python
y_pred = clf_2f.predict(X_test)
accuracy_score(Y_test, y_pred)
```

```python
clf_2f.feature_importances_
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(data=df, x='flavanoids', y="color_intensity", hue= "label")
```

```python
plt.figure(figsize=(16, 8))
axes=[0, 5.5, 0, 13]
x1s = np.linspace(axes[0], axes[1], 100)
x2s = np.linspace(axes[2], axes[3], 100)
x1, x2 = np.meshgrid(x1s, x2s)
X_new = np.c_[x1.ravel(), x2.ravel()]
y_pred = clf_2f.predict(X_new).reshape(x1.shape)
plt.contourf(x1, x2, y_pred, alpha=0.3)
sns.scatterplot(data=df, x='flavanoids', y="color_intensity", hue= "label")
plt.figtext(.5,.9,'Simple Model Graph', fontsize=25, ha='center')
```

```python
clf_2f_overfit = DecisionTreeClassifier(criterion='gini', max_depth=10, random_state=42)
clf_2f_overfit.fit(X_train, Y_train)

plt.figure(figsize=(16, 8))
axes=[0, 5.5, 0, 13]
x1s = np.linspace(axes[0], axes[1], 100)
x2s = np.linspace(axes[2], axes[3], 100)
x1, x2 = np.meshgrid(x1s, x2s)
X_new = np.c_[x1.ravel(), x2.ravel()]
y_pred = clf_2f_overfit.predict(X_new).reshape(x1.shape)
plt.contourf(x1, x2, y_pred, alpha=0.3)
plt.figtext(.5,.9,'Overfit Model Graph', fontsize=25, ha='center')
sns.scatterplot(data=df, x='flavanoids', y="color_intensity", hue= "label")
```

```python
acc_2f_overfit = accuracy_score(clf_2f_overfit.predict(X_test),Y_test)
print('max_depth = 10 에 대한 정확도 스코어는 {}입니다'.format(acc_2f_overfit))
```
